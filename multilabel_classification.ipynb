# Data Preparation 
Tamil grammar error annotated corpus
We prepare our dataset into 


validation set(labelled)

1.   train set(labelled) - To be used for training models
2.   test set(labelled) - To be used for performance  
3.   validation set(labelled) - To be used for validation(hyperparameter optimization)

Dataset contains 7 columns.
1. grammar_text
2. error_text
3. right_tag - PNG tag
4. tag
5. label1 - Person(1,2,3)
6. label2 - Number(sg - singular, pl - plural)
7. label3 - Gender(m - male, female - f,  n-, he-, e-, o - null value)

```python
import pandas as pd
train_df = pd.read_csv('your_file_path',
                       delimiter='\t', 
                       header=None, 
                       names=['grammar text','error text', 'right_tag','tag','label1','label2','label3'])
```

```python
train_df.head
```

### Filling NaN values for labels
Error dataset contains null values for some labels. Fill that null values to some special characters('o').


```python
train_preprocess = train_df.fillna("o")
```

###Encoding
Convert labels into onehot encoding.

```python
train_preprocess
```

```python
one_hot_encoded_data = pd.get_dummies(train_preprocess, columns = ['label1', 'label2','label3'])
```

```python
one_hot_encoded_data.head(10)
```

```python
one_hot_encoded_data['labels'] = list(zip(one_hot_encoded_data.label1_1.tolist(),
                                          one_hot_encoded_data.label1_2.tolist(),
                                          one_hot_encoded_data.label1_3.tolist(),
                                          one_hot_encoded_data.label2_sg.tolist(),
                                          one_hot_encoded_data.label2_pl.tolist(),
                                          one_hot_encoded_data.label3_e.tolist(),
                                          one_hot_encoded_data.label3_f.tolist(),
                                          one_hot_encoded_data.label3_m.tolist(), 
                                          one_hot_encoded_data.label3_he.tolist(),
                                          one_hot_encoded_data.label3_n.tolist() ,
                                          one_hot_encoded_data.label3_o.tolist() 
                                          ))
one_hot_encoded_data['text'] = one_hot_encoded_data['error text'].apply(lambda x: x.replace('\n', ' '))

```

```python
df = one_hot_encoded_data[['text','labels']]
```

```python
df.head
```

```python
new_df = df.sample(50000,random_state=4)
```

```python
from sklearn.model_selection import train_test_split


train_df, eval_df = train_test_split(new_df, test_size=0.2,random_state=4)
eval_df , test_df = train_test_split(eval_df,test_size=0.5,random_state=4)


```

###Training a Model with sensible Defaults
Two of the most critical hyperparameters to consider when training a Transformer model on an NLP task is the **learning rate** and the **number of training epochs**.

learning rate --> 5e-5 (1e-5)

training epochs --->2 or 3

train_batch_size -->2

gradient_accumulation_steps --->16

max_seq_length --> 512



```python
import logging
from statistics import mean

import pandas as pd
from sklearn.metrics import accuracy_score


from simpletransformers.classification import  ClassificationArgs, MultiLabelClassificationModel
```

```python
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
from simpletransformers.classification import MultiLabelClassificationModel

#Create a TranformerModel
model = MultiLabelClassificationModel('albert', 'ai4bharat/indic-bert',
                                      args={
        "eval_batch_size" : 8,
"evaluate_during_training" : True,
"evaluate_during_training_silent" : False,
"evaluate_during_training_steps" : -1,
"save_eval_checkpoints" : False,
"save_model_every_epoch" : False,
"learning_rate" : 0.00004202025664941036,
"manual_seed" : 4,
"max_seq_length" : 512,
"multiprocessing_chunksize" : 5000,
"no_cache" : True,
"num_train_epochs": 6,
"overwrite_output_dir" : True,
"reprocess_input_data" : True,
"train_batch_size" : 2,
"gradient_accumulation_steps" : 16,
"labels_list" : ["1-person","2-person","3-person","singular","plural","e","f","m","he","n","o"],
"output_dir" : "default_output",
"best_model_dir" : "default_output/best_model"},
                                      use_cuda=True,
                                      num_labels=11)
```

```python
# Train the model
model.train_model(train_df,
                  eval_df=eval_df)
```

###Basic Hyperparameter Optimization
W&B Sweeps integration in simple transformers simplifies the process of conducting hyperparameter optimization.

we will focus on the same two hyperparameters as in the earlier section, i.e. learning rate and the number of training epochs.

```python
import wandb
logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

```python
args={"eval_batch_size" : 8,
"evaluate_during_training" : True,
"evaluate_during_training_silent" : False,
"evaluate_during_training_steps" : -1,
"save_eval_checkpoints" : False,
"save_model_every_epoch" : False,
"manual_seed" : 4,
"max_seq_length" : 512,
"multiprocessing_chunksize" : 5000,
"no_cache" : True,
"overwrite_output_dir" : True,
"reprocess_input_data" : True,

"gradient_accumulation_steps" : 16,
"labels_list" : ["1-person","2-person","3-person","singular","plural","e","f","m","he","n","o"],
"output_dir" : "default_output",
"best_model_dir" : "default_output/best_model",
"wandb_project" : "TamGramCheck - Hyperparameter Optimization"}
```

```python
sweep_config = {
    "name": "vanilla-sweep-batch-16",
    "method": "bayes",
    "metric": {"name": "LRAP", "goal": "maximize"},
    "parameters": {
        "num_train_epochs": {"min": 1, "max": 6},
        "learning_rate": {"min": 1e-5, "max": 5e-5},
        "train_batch_size":{"min":2,"max":16}
    },
    "early_terminate": {"type": "hyperband", "min_iter": 6,},
}
```

```python

sweep_id = wandb.sweep(sweep_config, project="TamGramCheck - Hyperparameter Optimization-Final")
```

```python
import wandb

def train():
    # Initialize a new wandb run
    wandb.init(resume=True)

    # Create a TransformerModel
    model = MultiLabelClassificationModel('albert', 'ai4bharat/indic-bert',
        use_cuda=True,
        num_labels=11,
        args=args,
        sweep_config=wandb.config,
    )

    # Train the model
    model.train_model(
        train_df,
        eval_df=eval_df
    )

    # Sync wandb
    wandb.join()
```

```python
wandb.agent(sweep_id, train)
```
